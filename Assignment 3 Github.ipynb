# DataSet
This dataset consists of 2 files "hashtag_donaldtrump , hashtag_joebiden" which are (410 , 363) MB respectively so i took the first 100 rows of each file as a sample and put them in a new .csv file name "Trump , joebiden" to save time and to be able to upload them.

Link of the dataset : 'https://www.kaggle.com/manchunhui/us-election-2020-tweets'

```python
import pandas as pd
import re
#nltk.download()                       #In case you don't have nltk corpus Uncomment it and download them.
from nltk import word_tokenize , LancasterStemmer , pos_tag 
import string
import nltk
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.lancaster import LancasterStemmer
from nltk.corpus import stopwords
```

```python
Trump    = pd.read_csv('Trump.csv' , encoding = "ISO-8859-1")
joebiden = pd.read_csv('joebiden.csv' , encoding = "ISO-8859-1")
```

```python
Trump.columns
```

```python
Trump    = Trump.iloc[:,0 : 3]
joebiden = joebiden.iloc[:,0 : 3] 
```

```python
Trump.head()
```

```python
def Preprocessing(df_name):
    
    stop_words = set(stopwords.words('english')) 
    stemmer = LancasterStemmer()
    
    df_name = df_name.assign(tokenize = [word_tokenize(word) for word in df_name['tweet']])
    df_name = df_name.assign(clean_text = [(re.sub('[%s]' % re.escape(string.punctuation), ' ' , (word))) for word in df_name['tweet']])
    df_name = df_name.assign(lower_text = [word.lower() for word in df_name['clean_text']])
    
        
    df_name['text_without_stop_words'] = df_name['lower_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
    df_name['stemmed_text'] = df_name['text_without_stop_words'].apply(lambda x: [stemmer.stem(y) for y in x.split()])
    df_name['POS_text'] = df_name['stemmed_text'].apply(lambda x: [pos_tag(word_tokenize(word)) for word in x])
    return df_name
```

```python
cleaned_Trump = Preprocessing(Trump)
cleaned_joebiden = Preprocessing(joebiden)
```
